```python
n = int(input())
groups = list(map(int, input().split()))

cnt = [0] * 5
for g in groups:
    cnt[g] += 1

taxis = 0

taxis += cnt[4]

taxis += cnt[3]
cnt[1] = max(0, cnt[1] - cnt[3])  

taxis += cnt[2] // 2
if cnt[2] % 2 == 1:  
    taxis += 1
    cnt[1] = max(0, cnt[1] - 2)  

taxis += (cnt[1] + 3) // 4

print(taxis)
```
