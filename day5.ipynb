```python
t=int(input())

for _ in range(t):
    n=int(input())
    if n%3 == 0:
        print("Second")
    elif n%3==1:
        print("First")
    elif n%3==2:
        print("First")
```

```python
n, l = map(int, input().split())
a = list(map(int, input().split()))

a.sort()

max_dist = a[0]

for i in range(1, n):
    max_dist = max(max_dist, (a[i] - a[i-1]) / 2)

max_dist = max(max_dist, l - a[-1])

print(max_dist)
```

```python
n = int(input())
cards = list(map(int, input().split()))

left = 0
right = n - 1

sereja = 0
dima = 0

turn = 0  

while left <= right:
    if cards[left] > cards[right]:
        value = cards[left]
        left += 1
    else:
        value = cards[right]
        right -= 1

    if turn == 0:
        sereja += value
    else:
        dima += value

    turn = 1 - turn

print(sereja, dima)
```

```python
k, r = map(int, input().split())
for n in range(1, 11):
    total = n * k
    if total % 10 == 0 or (total - r) % 10 == 0:
        print(n)
        break
```
