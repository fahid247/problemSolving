```python
raw_numbers=input()
k,n,w=map(int,raw_numbers.split(" "))
total_cost = 0
for i in range(1, w + 1):
    total_cost += i * k
    
if total_cost>n:
    borrow=total_cost-n
    print(borrow)

else:
    print("0")
```
