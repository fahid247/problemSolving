```python
n, t = map(int, input().split())
s = list(input())

for _ in range(t):
    i = 0
    while i < n - 1:
        if s[i] == "B" and s[i+1] == "G":
            s[i], s[i+1] = s[i+1], s[i]
            i += 2
        else:
            i += 1

print("".join(s))
```

```python
s=input()
target="hello"
j=0
for c in s:
    if j<len(target) and c==target[j]:
        j+=1
if j==len(target):
        print("YES")
else:
        print("NO")
```

```python
n = int(input())
count = 0

for _ in range(n):
    p, q = map(int, input().split())
    if q - p >= 2:
        count += 1

print(count)
```

```python
n = int(input())

total = 0

for i in range(1, n+1):
    if i % 2 == 0:  
        total += i
    else:           
        total -= i

print(total)
```

```python
n = int(input())

count = 0

for _ in range(n):
    p, v, t = map(int, input().split())

    if p + v + t >= 2:
        count += 1

print(count)
```
