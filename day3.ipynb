```python
n = int(input())
a = list(map(int, input().split()))

max_height = max(a)
min_height = min(a)

max_pos = a.index(max_height)
min_pos = n - 1 - a[::-1].index(min_height)

moves = max_pos + (n - 1 - min_pos)

if max_pos > min_pos:
    moves -= 1

print(moves)
```

```python
a = int(input())
b = int(input())
c = int(input())

max_value = max(
    a + b + c,
    a * b * c,
    (a + b) * c,
    a * (b + c)
)

print(max_value)
```

```python
k = int(input())
l = int(input())
m = int(input())
n = int(input())
d = int(input())

damaged = 0

for i in range(1, d + 1):
    if i % k == 0 or i % l == 0 or i % m == 0 or i % n == 0:
        damaged += 1

print(damaged)
```
