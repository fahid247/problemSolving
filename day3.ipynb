```python
n = int(input())

x_levels = list(map(int, input().split()))
y_levels = list(map(int, input().split()))

levels = set(x_levels[1:] + y_levels[1:])

if len(levels) == n:
    print("I become the guy.")
else:
    print("Oh, my keyboard!")
```

```python
n = int(input())
a = list(map(int, input().split()))

max_height = max(a)
min_height = min(a)

max_pos = a.index(max_height)
min_pos = n - 1 - a[::-1].index(min_height)

moves = max_pos + (n - 1 - min_pos)

if max_pos > min_pos:
    moves -= 1

print(moves)
```
